# Image Scaler

#### This notebook resizes or rescales an image to a specified size.

## Importing the libraries

```python
import matplotlib.pyplot as plt
from tkinter import Tk
from tkinter.filedialog import askopenfilename
from PIL import Image
from torchvision import transforms
```

## Open file dialog to select an image

```python
root = Tk()
file_path = askopenfilename(title="Select an image")
print("Selected file:", file_path)
```

```python
img = Image.open(file_path)
img
```

```python
copy_img = img.copy()
```

```python
transforms.ToTensor()(copy_img).shape
```

## Desired height and width

```python
height, width = int(input("desired height: ")), int(input("desired width: "))
gray_scale = input("Do you want to have the image in grayscale? [y/n]: ")
if gray_scale.lower() == 'n':
    data_transform = transforms.Resize(size=(height, width))
else:
    data_transform = transforms.Compose([
        transforms.Resize(size=(height, width)),
        transforms.Grayscale(num_output_channels=1)
])
```

```python
transformed_img = data_transform(copy_img)
transformed_img
```

## Saving the transformed image

```python
transformed_img.save("Transformed.jpg")
```
